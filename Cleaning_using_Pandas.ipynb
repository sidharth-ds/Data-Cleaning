## Drop( ):

```python
df4 = df.dropna()
df4
```

* only when all the values in a row are null, that row is dropped:

```python
df5 = df.dropna(how='all')
df5
```

* if any value in a row is null, then that row is dropped:

```python
df6 = df.dropna(how='any')  # same as df.dropna()
df6
```

## subset= 
* to call only the particular column/columns

```python
df7 = df.dropna(subset=["hsc"])
df7
```

```python
df8 = df.dropna(subset=["hsc","SSLC"])
df8
```

Dropping via column-wise:
* axis="columns"
* axis=1

```python
df9 = df.dropna(axis="columns")
df9
```

## To check the Duplicate rows:

```python
df2
```

```python
df10 = df2.duplicated()
df10
```

by default:
* first is original, rest are all taken as duplicates
* keep = first (default)

keep = last
* last is kept as original
* rest are all taken as  duplicates     
                         


```python
df11 = df2.duplicated(keep='last')
df11
```

```python
df12 = df2.duplicated(keep='first')  # dy default
df12
```

keep=False:
* all the rows having same values are taken as duplicates including the original one.

```python
df13 = df2.duplicated(keep=False) 
df13
```

```python
df14 = df2.drop_duplicates()
df14
```

```python
df15 = df2.drop_duplicates(keep=False)
df15
```

```python
df16 = df2.drop_duplicates(subset=["college"])
df16
```

```python
df17 = df2.drop_duplicates(subset=["college"],keep=False)
df17
```
